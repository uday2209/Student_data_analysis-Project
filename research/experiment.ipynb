```python
all_scores = pd.melt(df, id_vars=[],value_vars=["math_score", "reading_score", "writing_score"],var_name="subject", value_name="score")
```

```python
all_scores["subject"] = all_scores["subject"].str.replace("_score", "")
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=all_scores, x="score", hue="subject", multiple="stack", bins=10)
plt.title("Histogram of Scores")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```
